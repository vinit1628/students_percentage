```python
# Importing necessary libraries for the project



import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Dataset Path


url = "http://bit.ly/w-data"
df=pd.read_csv(url)
```

```python
df.head()
```

```python
df.info()
```

```python
#using scatterplot for visualizations

plt.figure(figsize=(3,3),dpi=200)
sns.scatterplot(x='Hours',y='Scores',data=df)
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.title('Hours vs Percentage')
```

```python
#telling the model about the dependent and independent variable
X = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
```

```python
#training and testing the data


from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
df = pd.DataFrame({"Actual": y_test,"Predicted":y_pred})
```

```python
df
```

```python
# Evaluating the performance

from sklearn import metrics
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
mean_absolute_error(y_test,y_pred)

```

```python
r2_score(y_test,y_pred)
```

# what will be the predicted score if the student studies for 9.25/hours a day?

```python
X_new = np.array(9.25).reshape(-1,1)
y_pred = model.predict(X_new)
```

```python
print("predicted score = {}".format(y_pred[0]))
```
